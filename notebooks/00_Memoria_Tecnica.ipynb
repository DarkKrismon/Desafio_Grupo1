# Memoria del trabajo

## Guia del procesamiento de los datos

1. Molde de datos en crudo que coincida (.json ya que eso le llega a la API)

2. Interferencias que no cuadran -> señalar PELIGRO.

3. Train y Test

4. Trasformación (Nulos, Signos raros...)

5. Label Encoding con diccionarios (Trasforman string -> num)

6. Standart Scaler (cambia num grandes a pequeños)

7. Añadir pesos a la clases importantes.

8. Definir límites para evitar Outliners

9. Pasarlo por el modelo y sacar esa predict esperada

10. Comprimir datos importantes y pasarlos a la API

## Guia del procesamiento de ML

1. Random Forest o XGBoost/Gradient Boosting

2. Devuelve `predict_proba()` que aporta el porcentaje de riesgo

3. Evaluar y optimizar mirando el RECALL y el F1-SCORE

4. Importante exportar el modelo `.pkl` con los traductores de `StandardScaler` y de `LabelEncoder`

Podemos tratar el step para que indique la hora del día de las transacciones y poder detectar fraude con mayor facilidad.
